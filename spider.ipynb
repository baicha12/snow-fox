# XPATH语法
## 一层一层往下挖模式
> - '/'	从根节点选取
- '//'	从匹配选择的当前节点选择文档中的节点，而不考虑它们的位置 **
- '.'	选取当前节点 **
- '..'	选取当前节点的父节点
- '@'   提取属性和控制属性 ** 
- '[]'  属性容器
- '*'	匹配任何元素节点
- text() 以文本的方式返回提取的内容 **

```python
html = '''
<html>

<head>
<title>我的第一个 HTML 页面</title>
</head>

<body>
<p>body 元素的内容会显示在浏览器中。</p>
<p>title 元素的内容会显示在浏览器的标题栏中。</p>
<div class="hahah">今天天气真好</div>
<div class="lalal">今天天气真不好</div>
<a href="http://www.baidu.com">百度一下</a>

<div class="Joker">
    <a href="http://www.baidu.com">百度一下</a>
    <a href="http://www.baidu.com">百度一下</a>
    <a href="http://www.baidu.com">百度一下</a>
    <a href="http://www.baidu.com">百度一下</a>
</div>
</body>

</html>
'''
```

```python
from lxml import etree
```

```python
tree = etree.HTML(html) # 将网页源代码转换成Xpath能看到懂的东西
```

```python
tree.xpath('//body/p[2]/text()')
```

```python
tree.xpath('//body/div[@class="lalal"]/text()')
```

```python
tree.xpath('//body/a/@href')
```

```python
a_xpath=tree.xpath('//div[@class="Joker"]/a/@href')
a_xpath
```

```python
import time
a_xpath=tree.xpath('//div[@class="Joker"]/a')
a_xpath
for i in a_xpath:
    href=i.xpath('./@href')
    time.sleep(1)
    print(href)
```

# 实战2

```python
import requests
from lxml import etree

file = '/Users/huwang/Desktop/day04/wenben.txt' # 定义一个文件路径，使得可以储存
open_file = open(file,mode='a',encoding='utf8') # 以append 追加的方式打开文件
header = {
   'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36' 
}  # 设置浏览器信息，伪造浏览器
response = requests.get('http://www.17k.com/list/2839638.html',headers = header) # 进行请求，并携带浏览器信息
response.encoding = 'utf8' # 转换网页的编码，不是任何时候都需要，当源代码出现乱码的时候才需要
html=response.text # 获取网页源代码
tree = etree.HTML(html) # 将其转化成xpath
a_xpath=tree.xpath('//div[@class="Main List"]/dl[@class="Volume"]/dd/a') # 获取dd下面所有的a
start_url = 'http://www.17k.com' # 初始化url，目的是为了进行下面的完整网址的拼接
for i in a_xpath:
    url = i.xpath('./@href')[0] # 通过xpath 获取不完整的url（href）
    all_url = start_url+url # 进行路径（网址）的拼接
    response2 = requests.get(all_url) # 继续请求每个章节
    response2.encoding = 'utf8'
    html2=response2.text # 获取章节内容的源代码
    tree2 = etree.HTML(html2) 
    p_xpath=tree2.xpath('//*[@id="readArea"]/div[1]/div[2]/text()')  # 获取章节内容的文本
    open_file.write(str(p_xpath))  # 写入到本地
open_file.close() # 当for 循环执行完毕，关闭文件
```
